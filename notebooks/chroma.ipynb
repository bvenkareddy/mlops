```python
from chromadb.config import Settings
client = chromadb.PersistentClient(path="C:\\Users\\theco\\OneDrive\\Documents\\git_repo\\mlops\\chromadb",settings=Settings(allow_reset=True))
```

```python
client.heartbeat()
client.reset()
```

## Create Collection

```python
collection = client.get_or_create_collection(name="test_collection",metadata={"hnsw:space": "cosine"})
```

## Add Raw documents to collection

```python
collection.add(
    documents=["C:\\Users\\theco\\OneDrive\\Documents\\git_repo\\mlops\\data\\pyspark.pdf"],
    metadatas=[{"text":"Intro to pyspark"}],
    ids=["spark_001"]
)
```
